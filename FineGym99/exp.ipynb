```python
from pathlib import Path

import cv2


video_dir = Path("./FineGym99")

for video_file in video_dir.glob("*.mp4"):
    cap = cv2.VideoCapture(str(video_file))
    cap.set(cv2.CAP_PROP_POS_FRAMES, 100)
    print(cap.get(cv2.CAP_PROP_FPS))
    print(video_file.stem)
```

```python
error_video = Path("./FineGym99/OIJe8fQw7Wo.2018 U.S. Gymnastics Championships - Women - Day 1 - Olympic Channel Broadcast.mp4")

cap = cv2.VideoCapture(str(error_video))
print(cap.get(cv2.CAP_PROP_FPS))
cap.set(cv2.CAP_PROP_POS_FRAMES, 1)
print(error_video.name)
```
